```python
import torch
from transformers import pipeline
from PIL import Image

# 1. モデルのダウンロード (DA3-Small)
model = pipeline(task="depth-estimation", model="LiheYoung/depth-anything-small-hf", device=0, trust_remote_code=True)
```

```python
# 2. 画像ファイルのアップロード（実行するとボタンが出ます）
from google.colab import files
uploaded = files.upload()
img_path = list(uploaded.keys())[0]  # アップロードした1枚目を取得

# 3. 深度画像の推論
image = Image.open(img_path)
result = model(image)
depth = result["depth"]
```

```python
# 4. 画像を並べて表示
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis("off")

#axes[1].imshow(depth, cmap="inferno") # cmap="gray" にすると白黒になります
axes[1].imshow(depth, cmap="gray") # cmap="gray" にすると白黒になります
axes[1].set_title("Depth Map (farther is darker)")
axes[1].axis("off")

plt.show()
```

```python
# 深度画像をダウンロードしたいとき
filename = "depth_map.png"
result["depth"].save(filename)
from google.colab import files
files.download(filename)
```
